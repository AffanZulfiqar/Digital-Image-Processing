# **TASK 01 : Traffic Sign Detection**

**Description:**

In this Task I loaded the traffic sign image and converted it to grayscale. Then, I applied thresholding to separate regions and used edge detection to highlight boundaries. Finally, I removed noise and displayed all results in a 2x2 grid with titles.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "/content/traffic_sign.jpeg"
image = cv2.imread(image_path)

# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding
_, thresholded = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Edge Detection
edges = cv2.Canny(thresholded, 50, 150)

# Remove Noise and Extract Features
kernel = np.ones((5,5), np.uint8)
processed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

titles = ["Grayscale Image", "Thresholded Image", "Edge Detection", "Processed Image"]
images = [gray, thresholded, edges, processed]

plt.figure(figsize=(10, 10))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```

# **Task 02 : Satellite Imaging**

**Description:**

In task 2 I blended satellite images and converted the result to grayscale. Then, I applied thresholding to segment land and water, performed edge detection to highlight boundaries, and created a mask to focus on a specific region. At last, I displayed all processed images.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img1 = cv2.imread("/content/satellite.png")
img2 = cv2.imread("/content/satellite.png")

# Blend images
blend = cv2.addWeighted(img1, 0.6, img2, 0.4, 0)

gray = cv2.cvtColor(blend, cv2.COLOR_BGR2GRAY)

# Thresholding
_, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

# Edge detection
edges = cv2.Canny(thresh, 50, 150)

# Mask for specific region
mask = np.zeros_like(gray)
cv2.rectangle(mask, (100, 100), (300, 300), 255, -1)
roi = cv2.bitwise_and(blend, blend, mask=mask)
cv2_imshow(blend)
```

# **TASK 03: Smart Surveillance System**

**Description:**


I processed the image in this task first to convert it to grayscale and apply thresholding to create a binary mask. Then, I filtered noise using morphological operations and detected motion using image subtraction. At last, I displayed all results for comparison.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "/content/task_3.jpeg"
image = cv2.imread(image_path)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding to create a binary mask
_, binary_mask = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

# Logical AND & OR operations to filter noise
kernel = np.ones((5,5), np.uint8)
filtered_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_OPEN, kernel)

image2 = cv2.GaussianBlur(gray, (5,5), 0)

# Detect motion using image subtraction
motion = cv2.absdiff(gray, image2)

titles = ["Binary Mask", "Filtered Mask", "Motion Detection"]
images = [binary_mask, filtered_mask, motion]

plt.figure(figsize=(15, 7))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i], fontsize=14)
    plt.axis("off")
plt.show()
```

**Conclusion:**

In this lab, I blended satellite images and applied a mask to highlight specific regions. I used thresholding and edge detection to separate land and water while extracting key features. Noise was filtered using morphological operations, and motion detection was performed using image subtraction.
