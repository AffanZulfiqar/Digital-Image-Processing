# **Scenario no 01 :** Enhancing X-ray Images for Medical Diagnosis

**Explanation:**

In this task, I have removed noise using Non-Local Means Denoising, enhanced contrast with CLAHE, and sharpened details using Unsharp Masking.At last I have detected edges with Canny and overlaid them on the enhanced image for better visualization.

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt

image = cv2.imread('/content/Scanerio 01.png', cv2.IMREAD_GRAYSCALE)
denoised_image = cv2.fastNlMeansDenoising(image, h=10, templateWindowSize=7, searchWindowSize=21)

gaussian_blur = cv2.GaussianBlur(denoised_image, (0, 0), 3)
unsharp_image = cv2.addWeighted(denoised_image, 1.5, gaussian_blur, -0.5, 0)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
enhanced_image = clahe.apply(unsharp_image)
edges = cv2.Canny(enhanced_image, threshold1=50, threshold2=150)
color_edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
overlay_image = cv2.addWeighted(cv2.cvtColor(enhanced_image, cv2.COLOR_GRAY2BGR), 0.8, color_edges, 0.2, 0)

cv2.imwrite('denoised_xray.png', denoised_image)
cv2.imwrite('enhanced_xray.png', enhanced_image)
cv2.imwrite('edges_xray.png', edges)
cv2.imwrite('overlay_xray.png', overlay_image)
plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(denoised_image, cmap='gray')
plt.title('Denoised Image')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(enhanced_image, cmap='gray')
plt.title('Enhanced Image')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(edges, cmap='gray')
plt.title('Edges (Canny)')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(overlay_image, cv2.COLOR_BGR2RGB))
plt.title('Overlay Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario no 02 :** Object Detection in Autonomous Vehicles

**Explanation:**

In this task, I have enhanced contrast using Histogram Equalization, smoothed the image with Median Blur to remove noise, and detected edges using Canny. I have applied Bilateral Filtering to reduce noise while preserving important edge details.

```python
image_path = "/content/Scanerio 02.png"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

equalized = cv2.equalizeHist(image)
smoothed = cv2.medianBlur(equalized, 5)
edges = cv2.Canny(smoothed, 50, 150)
bilateral_filtered = cv2.bilateralFilter(edges, 9, 75, 75)

plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(equalized, cmap='gray')
plt.title('Histogram Equalized')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(smoothed, cmap='gray')
plt.title('Smoothed Image (Median Blur)')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(edges, cmap='gray')
plt.title('Edges (Canny)')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(bilateral_filtered, cmap='gray')
plt.title('Bilateral Filtered')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario 03:** Restoring Old and Blurry Photographs

**Explanation:**

In this task, I have removed noise using Non-Local Means Denoising, smoothed the image with Gaussian Blur, and enhanced details using a sharpening filter. Finally, I have restored the image by combining the sharpened and smoothed versions for a balanced enhancement.

```python
image_path = "/content/Scanerio 03.png"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

denoised = cv2.fastNlMeansDenoising(image, None, 10, 7, 21)
gaussian_blur = cv2.GaussianBlur(denoised, (3, 3), 0)
sharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened = cv2.filter2D(gaussian_blur, -1, sharpen_kernel)
final_restored = cv2.addWeighted(sharpened, 1.5, gaussian_blur, -0.5, 0)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(denoised, cmap='gray')
plt.title('Denoised Image')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Smoothed Image')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(final_restored, cmap='gray')
plt.title('Final Restored Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario 04:** Enhancing Fingerprint Images for Biometric Systems

**Explanation:**

In this task, I have enhanced contrast using Histogram Equalization, smoothed out noise with Gaussian Blur and Bilateral Filtering, and sharpened the fingerprint ridges for better clarity. Finally, I have detected edges using Canny to highlight key features for biometric recognition.

```python
image_path = "/content/Scanerio 04.png"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

equalized = cv2.equalizeHist(image)
blurred = cv2.GaussianBlur(equalized, (3, 3), 0)
filtered = cv2.bilateralFilter(blurred, 9, 75, 75)

sharpen_kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpened = cv2.filter2D(filtered, -1, sharpen_kernel)

edges = cv2.Canny(sharpened, 50, 150)

plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalized")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(blurred, cmap='gray')
plt.title("Gaussian Blurred")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(filtered, cmap='gray')
plt.title("Bilateral Filtered")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
