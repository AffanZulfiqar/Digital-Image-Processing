# **Scenario no 01:** Bit-Plane Slicing for Document Scanning

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = '/content/Scanerio 01.png'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded correctly
if original_image is None:
    raise ValueError("Image not found. Please check the file path.")

# Extract bit planes
bit_planes = []
for i in range(8):
    bit_plane = (original_image >> i) & 1
    bit_planes.append(bit_plane * 255)

# Display all bit planes
plt.figure(figsize=(12, 8))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()

# Enhance the image using the higher bit planes (adjusting weights)
enhanced_image = (bit_planes[5] * 1.0 + bit_planes[6] * 2.0 + bit_planes[7] * 4.0)

# Normalize the enhanced image to 0-255
enhanced_image = cv2.normalize(enhanced_image, None, 0, 255, cv2.NORM_MINMAX)
enhanced_image = enhanced_image.astype(np.uint8)

# Apply thresholding to enhance contrast
_, enhanced_image = cv2.threshold(enhanced_image, 100, 255, cv2.THRESH_BINARY)

# Compare the original image and the enhanced image
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Document Image')
plt.axis('off')

# Enhanced image
plt.subplot(1, 2, 2)
plt.imshow(enhanced_image, cmap='gray')
plt.title('Enhanced Document Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario no 02 :** Histogram Processing for Medical Imaging

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the medical image (grayscale)
image_path = "/content/Scanerio 02.jpg"  # Update with correct path
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded correctly
if original_image is None:
    raise ValueError("Error: Could not load image. Please check the file path.")

# Function to plot histogram
def plot_histogram(image, title):
    plt.figure(figsize=(6, 4))
    plt.hist(image.ravel(), bins=256, range=[0, 256], color='gray', alpha=0.7)
    plt.title(title)
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.show()

# Function to check exposure status
def check_exposure(image):
    """Check if the image is underexposed, overexposed, or well-balanced."""
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
    total_pixels = image.size

    # Calculate thresholds
    dark_pixels = np.sum(histogram[:50]) / total_pixels
    bright_pixels = np.sum(histogram[200:]) / total_pixels

    if dark_pixels > 0.5:
        return "Underexposed"
    elif bright_pixels > 0.5:
        return "Overexposed"
    else:
        return "Well-Balanced"

# Function to enhance image contrast
def enhance_image(image, method="hist_eq"):
    """Enhance image using histogram equalization or contrast stretching."""
    if method == "hist_eq":
        enhanced = cv2.equalizeHist(image)
    elif method == "stretch":
        min_val, max_val = np.percentile(image, (2, 98))
        enhanced = np.clip((image - min_val) * (255 / (max_val - min_val)), 0, 255).astype(np.uint8)
    else:
        raise ValueError("Invalid method. Choose 'hist_eq' or 'stretch'.")
    return enhanced

# Identify exposure status
exposure_status = check_exposure(original_image)
print(f"Exposure Status: {exposure_status}")

# Enhance the image
if exposure_status != "Well-Balanced":
    enhanced_image = enhance_image(original_image, method="hist_eq")
else:
    enhanced_image = original_image  # No enhancement needed

# Plot results
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Medical Image")
plt.axis('off')

# Histogram of Original Image
plt.subplot(1, 3, 2)
plot_histogram(original_image, "Histogram of Original Image")

# Enhanced Image
plt.subplot(1, 3, 3)
plt.imshow(enhanced_image, cmap='gray')
plt.title("Enhanced Image (Histogram Equalization)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario 03:** Global Histogram Equalization for Satellite Imaging

```python
image_path = '/content/Scanerio 03.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_COLOR)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

gray_image = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

double_equalized_image = cv2.equalizeHist(equalized_image)
double_equalized_image_rgb = cv2.cvtColor(double_equalized_image, cv2.COLOR_GRAY2RGB)
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title('Original Satellite Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2RGB))
plt.title('First Equalized Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(double_equalized_image_rgb)
plt.title('Double Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario 04:** Transformation Functions for Night-Time Image Brightening

```python
# Load the dark night-time image in grayscale
image_path = "/content/Scanerio 04.jpg"  # Update path if needed
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded correctly
if original_image is None:
    raise ValueError(f"Image not found at path: {image_path}. Please check the file path.")

# Apply Logarithmic Transformation
# Convert image to float32 before applying log1p to handle potential large values
original_image = original_image.astype(np.float32)
log_transformed = cv2.normalize(np.log1p(original_image), None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

# ... rest of your code ...
# Apply Gamma Correction
gamma = 2.2  # Adjust gamma value as needed
gamma_corrected = np.array(255 * (original_image / 255) ** (1 / gamma), dtype=np.uint8)

# Compare the Original, Log Transformed, and Gamma Corrected images
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Night-Time Image")
plt.axis('off')

# Log Transformation
plt.subplot(1, 3, 2)
plt.imshow(log_transformed, cmap='gray')
plt.title("Logarithmic Transformation")
plt.axis('off')

# Gamma Correction
plt.subplot(1, 3, 3)
plt.imshow(gamma_corrected, cmap='gray')
plt.title(f"Gamma Correction (={gamma})")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Scenario 05:** Local Histogram Equalization for Wildlife Photography

```python
# Load the wildlife image (grayscale)
image_path = "/content/Scanerio 05.jpg"  # Update with correct path
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded correctly
if original_image is None:
    raise ValueError("Error: Could not load image. Please check the file path.")

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(original_image)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Wildlife Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(clahe_image, cmap='gray')
plt.title("Enhanced Image (CLAHE)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
